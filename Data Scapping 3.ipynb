```python
from bs4 import BeautifulSoup
import requests
```

```python
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_manufacturers_by_motor_vehicle_production'
page_to_scrap = requests.get(url)
soup = BeautifulSoup(page_to_scrap.text,'html')
print(soup)
```

```python
table = soup.findAll('table', attrs = {'class':'wikitable'})
table
```

```python
tables = table[0]
tables
```

```python
titles = tables.findAll('th')
titles
```

```python
all_titles = [title.text.strip() for title in titles]
all_titles
```

```python
df = pd.DataFrame(columns = all_titles)
```

```python
df
```

```python
column_data = tables.findAll('tr')
column_data
```

```python
for row in column_data[1:]:
    row_data = row.findAll('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
```

```python
for row in column_data[1:]:
    row_data = row.findAll('td')
    individual_row_data = [data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = individual_row_data
```
