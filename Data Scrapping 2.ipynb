```python
from bs4 import BeautifulSoup
import requests
```

```python
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
content = requests.get(url)
soup = BeautifulSoup(content.text,'html')
table = soup.find_all('table', attrs = {'class': 'wikitable'})
table
```

```python
tables = table[0]
tables
```

```python
titles = tables.find_all('th')
titles
```

```python
all_titles = [title.text.strip() for title in titles]
all_titles
```

```python
filtered_titles = all_titles[1:10]
filtered_titles
```

```python
df = pd.DataFrame(columns = filtered_titles)
df
```

```python
column_data = tables.find_all('tr')
column_data
```

```python
for row in column_data[2:]:
    row_data = row.find_all('td') 
    individual_row_data  = [data.text.strip() for data in row_data]
    print(individual_row_data)
    length = len(df)
    df.loc[length] = individual_row_data
```
